```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

NEE from LULC changes every year

East European forest steppe

```python
df1 = pd.read_excel('_/NEE_LULC_EEFS_byyears.xls', sheet_name=0)
print(df1.head())
```

```python
df1['MEAN'] = pd.to_numeric(df1['MEAN'], errors='coerce')
df1['TREND'] = pd.to_numeric(df1['TREND'], errors='coerce')
```

```python
pivotM = df1.pivot(index='TO', columns='FROM', values='MEAN').fillna(0)
pivotT = df1.pivot(index='TO', columns='FROM', values='TREND').fillna(0)
```

```python
colors = {
    'Agriculture': "#f6d607", 
    'Forest': "#197c0e", 
    'Grassland': "#f78e04",  
    'Wetland': "#0cbcf7",  
    'Settlement': "#ec2805", 
    'Other': "#c4bfbf" 
}

order_to = ['Agriculture', 'Forest', 'Grassland', 'Wetland', 'Settlement', 'Other']
order_from = ['Agriculture', 'Forest', 'Grassland', 'Wetland', 'Settlement', 'Other']

pivotM = pivotM.reindex(order_to)
pivotM = pivotM[order_from]

ax = pivotM.plot(kind='bar', stacked=True, figsize=(10,6),
                color=[colors[c] for c in pivotM.columns], edgecolor='none', zorder=2)

for container in ax.containers:
    labels = [f'{v.get_height():.2f}' if v.get_height() != 0 else '' for v in container]
    ax.bar_label(container, labels=labels, label_type='center', fontsize=8)

ymax = pivotM.sum(axis=1).max()
ymin = pivotM.sum(axis=1).min()
plt.ylim(ymin * 1.2, 0)

plt.ylabel('NEE (gC/m/year)')
plt.xlabel('Land Use/Cover Class in 2019', labelpad=10)
plt.xticks(rotation=0, ha='center') 
plt.title('Mean NEE in 2010-2019 for LULC transitions in East European forest steppe', pad=10)
plt.legend(title='From', bbox_to_anchor=(1.01,1.015))
plt.grid(color='gray', linestyle='-', linewidth=0.8, alpha=0.2, zorder=1)

labels = {
    'Agriculture': '-78.4', 
    'Forest': '-100.3', 
    'Grassland': '-89.0', 
    'Wetland': '-80.1',
    'Settlement': '-93.1',
    'Other': '-80.1'}

plt.text(0, ymin*1.05, 'Mean NEE', ha='center', va='top', fontsize=10)
for idx, label in enumerate(pivotM.index):
    plt.text(idx, ymin*1.09, labels[label],
             ha='center', va='top', fontsize=10)
plt.subplots_adjust(bottom=0.1)

plt.tight_layout()
plt.show()
```
